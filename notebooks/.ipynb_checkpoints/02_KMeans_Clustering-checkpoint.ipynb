```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# Load data
df = pd.read_csv('../data/customer_data.csv')

print("="*70)
print("K-MEANS CLUSTERING ANALYSIS")
print("="*70)
print(f"\nâœ… Loaded {len(df):,} customer records")

# Select features for clustering
features_for_clustering = [
    'Annual_Income', 
    'Spending_Score', 
    'Total_Purchases', 
    'Avg_Transaction_Value', 
    'Tenure_Months', 
    'Website_Visits_Monthly', 
    'Email_Open_Rate'
]

X = df[features_for_clustering]

print(f"\nðŸ“Š Features used for clustering: {len(features_for_clustering)}")
for i, feature in enumerate(features_for_clustering, 1):
    print(f"   {i}. {feature}")
```

```python
# Standardize the features (VERY IMPORTANT for K-Means!)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("="*70)
print("DATA STANDARDIZATION")
print("="*70)
print("âœ… Features standardized (mean=0, std=1)")
print(f"\nOriginal data shape: {X.shape}")
print(f"Scaled data shape: {X_scaled.shape}")

# Show example of standardization
print("\nðŸ“Š Before vs After Standardization (first 5 rows):")
comparison_df = pd.DataFrame({
    'Feature': features_for_clustering,
    'Original Mean': X.mean().values,
    'Original Std': X.std().values,
    'Scaled Mean': X_scaled.mean(axis=0),
    'Scaled Std': X_scaled.std(axis=0)
})
comparison_df
```

```python
# Elbow Method to find optimal number of clusters
print("\n" + "="*70)
print("FINDING OPTIMAL NUMBER OF CLUSTERS")
print("="*70)

inertias = []
silhouette_scores = []
K_range = range(2, 11)

print("\nTesting different values of K...")
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10, max_iter=300)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    sil_score = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(sil_score)
    print(f"K={k}: Inertia={kmeans.inertia_:,.2f}, Silhouette Score={sil_score:.4f}")

print("\nâœ… Analysis complete!")
```

```python
# Plot Elbow Curve and Silhouette Scores
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Elbow plot
ax1.plot(K_range, inertias, 'bo-', linewidth=2, markersize=10)
ax1.set_xlabel('Number of Clusters (K)', fontsize=13, fontweight='bold')
ax1.set_ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=13, fontweight='bold')
ax1.set_title('Elbow Method For Optimal K', fontsize=15, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axvline(x=5, color='red', linestyle='--', linewidth=2, label='K=5 (Optimal)')
ax1.legend()

# Silhouette plot
ax2.plot(K_range, silhouette_scores, 'ro-', linewidth=2, markersize=10)
ax2.set_xlabel('Number of Clusters (K)', fontsize=13, fontweight='bold')
ax2.set_ylabel('Silhouette Score', fontsize=13, fontweight='bold')
ax2.set_title('Silhouette Score Analysis', fontsize=15, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.axvline(x=5, color='red', linestyle='--', linewidth=2, label='K=5 (Optimal)')
ax2.legend()

plt.tight_layout()
plt.savefig('../visualization/elbow_method.png', dpi=300, bbox_inches='tight')
plt.show()
print("\nâœ… Saved: visualization/elbow_method.png")
```

```python
# Apply K-Means with optimal K=5
optimal_k = 5
print("\n" + "="*70)
print(f"APPLYING K-MEANS WITH K={optimal_k}")
print("="*70)

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10, max_iter=300)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Evaluation Metrics
silhouette_avg = silhouette_score(X_scaled, df['Cluster'])
davies_bouldin = davies_bouldin_score(X_scaled, df['Cluster'])

print(f"\nðŸ“Š Model Performance:")
print(f"   â€¢ Silhouette Score: {silhouette_avg:.4f} (Range: -1 to 1, higher is better)")
print(f"   â€¢ Davies-Bouldin Index: {davies_bouldin:.4f} (Lower is better)")

# Cluster distribution
print(f"\nðŸ“ˆ Cluster Distribution:")
cluster_counts = df['Cluster'].value_counts().sort_index()
for cluster_id, count in cluster_counts.items():
    percentage = (count / len(df)) * 100
    print(f"   â€¢ Cluster {cluster_id}: {count:,} customers ({percentage:.1f}%)")
```

```python
# Detailed Cluster Analysis
print("\n" + "="*70)
print("DETAILED CLUSTER PROFILES")
print("="*70)

cluster_summary = df.groupby('Cluster')[features_for_clustering].mean()

for cluster_id in range(optimal_k):
    count = cluster_counts[cluster_id]
    pct = (count / len(df)) * 100
    
    print(f"\nðŸŽ¯ CLUSTER {cluster_id}")
    print("-" * 70)
    print(f"Size: {count:,} customers ({pct:.1f}%)")
    print("\nAverage Characteristics:")
    
    for feature in features_for_clustering:
        value = cluster_summary.loc[cluster_id, feature]
        if feature == 'Annual_Income':
            print(f"   â€¢ {feature}: ${value:,.2f}")
        elif feature == 'Email_Open_Rate':
            print(f"   â€¢ {feature}: {value:.2%}")
        elif feature == 'Avg_Transaction_Value':
            print(f"   â€¢ {feature}: ${value:.2f}")
        else:
            print(f"   â€¢ {feature}: {value:.2f}")
```

```python
# PCA for 2D visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(14, 10))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], 
                     c=df['Cluster'], 
                     cmap='viridis', 
                     alpha=0.6, 
                     s=50,
                     edgecolors='black',
                     linewidth=0.5)

plt.colorbar(scatter, label='Cluster ID')
plt.xlabel(f'First Principal Component ({pca.explained_variance_ratio_[0]:.1%} variance)', 
           fontsize=13, fontweight='bold')
plt.ylabel(f'Second Principal Component ({pca.explained_variance_ratio_[1]:.1%} variance)', 
           fontsize=13, fontweight='bold')
plt.title('Customer Segments - K-Means Clustering\n(PCA 2D Projection)', 
          fontsize=16, fontweight='bold', pad=20)
plt.grid(True, alpha=0.3)

# Add cluster centers
centers_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], 
           c='red', marker='X', s=300, 
           edgecolors='black', linewidth=2,
           label='Cluster Centers')
plt.legend()

plt.tight_layout()
plt.savefig('../visualization/kmeans_clusters_pca.png', dpi=300, bbox_inches='tight')
plt.show()
print("\nâœ… Saved: visualization/kmeans_clusters_pca.png")
```

```python
# 3D Visualization
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(16, 12))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(df['Annual_Income'], 
                    df['Spending_Score'], 
                    df['Total_Purchases'], 
                    c=df['Cluster'], 
                    cmap='viridis', 
                    s=60, 
                    alpha=0.7,
                    edgecolors='black',
                    linewidth=0.5)

ax.set_xlabel('Annual Income ($)', fontsize=12, fontweight='bold', labelpad=10)
ax.set_ylabel('Spending Score', fontsize=12, fontweight='bold', labelpad=10)
ax.set_zlabel('Total Purchases', fontsize=12, fontweight='bold', labelpad=10)
ax.set_title('3D Customer Segmentation - K-Means Clustering', 
             fontsize=16, fontweight='bold', pad=20)

plt.colorbar(scatter, label='Cluster ID', shrink=0.5, pad=0.1)

plt.tight_layout()
plt.savefig('../visualization/kmeans_3d.png', dpi=300, bbox_inches='tight')
plt.show()
print("\nâœ… Saved: visualization/kmeans_3d.png")
```

```python
# Save clustered data
df.to_csv('../data/customer_data_clustered.csv', index=False)
print("\n" + "="*70)
print("K-MEANS CLUSTERING COMPLETE!")
print("="*70)
print(f"âœ… Clustered data saved to: data/customer_data_clustered.csv")
print(f"âœ… Total visualizations created: 3")
print(f"\nðŸŽ¯ Result: {optimal_k} distinct customer segments identified!")
```
